```python
# Convert 'MVT_DATE' to datetime
sales['MVT_DATE'] = pd.to_datetime(sales['MVT_DATE'])

# Calculate duration
date_min = sales['MVT_DATE'].min()
date_max = sales['MVT_DATE'].max()
duration = date_max - date_min

# Determine the exact time period
if duration.days < 31:
    time_period = f"{duration.days} days"
elif 31 <= duration.days < 61:
    time_period = f"{duration.days // 7} weeks"
elif 61 <= duration.days < 366:
    time_period = f"{duration.days // 30} months"
elif 366 <= duration.days < 731:
    time_period = "more than one year but less than two years"
elif 731 <= duration.days < 1096:
    time_period = "more than two years but less than three years"
else:
    time_period = "three years or more"

print(f"The data covers a period of: {time_period}")
```

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame (Replace this part with your actual DataFrame reading)

df = sales

# Function to generate historical data
def generate_historical_data(df, years_back=1):
    # Create an empty list to store the new rows
    new_data = []
    
    for year in range(1, years_back + 1):
        for _, row in df.iterrows():
            new_row = row.copy()
            # Update the MVT_DATE to one year back
            new_date = pd.to_datetime(row['MVT_DATE']) - pd.DateOffset(years=year)
            new_row['MVT_DATE'] = new_date.strftime('%Y-%m-%d')
            
            # Generate a new SALES_QTY following the same pattern
            sales_qty = int(row['SALES_QTY'])
            deviation = np.random.randint(-3, 3)  # Small deviation between -2 and 2
            new_sales_qty = max(1, sales_qty + deviation)  # Ensure new_sales_qty is at least 1
            new_row['SALES_QTY'] = new_sales_qty
            
            # Append the new row to the new_data list
            new_data.append(new_row)
    
    # Convert the new_data list to a DataFrame
    new_data_df = pd.DataFrame(new_data)
    
    # Combine the original DataFrame with the new_data DataFrame
#     combined_df = pd.concat([df, new_data_df], ignore_index=True)
    
    return new_data_df

# Generate the historical data
past_one_year = generate_historical_data(df)
past_two_year= generate_historical_data(past_one_year)
historical_df=pd.concat([df,past_one_year,past_two_year],ignore_index=True)

# Display the new DataFrame
print(past_one_year)
print(past_two_year)
```
